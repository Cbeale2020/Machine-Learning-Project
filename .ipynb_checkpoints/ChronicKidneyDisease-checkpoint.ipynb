Clean The Original Data

```python
# Dependencies
import pandas as pd
```

```python
# Load in file
original_data = "Data/unclean_data.csv"
```

```python
# Read and display the dataset
original_data_df = pd.read_csv(original_data)
original_data_df.head()
```

```python
# list all columns from the original dataset 
#list(original_data_df.columns) 
```

```python
# rename some columns from the original dataset for clarity
original_data_df.rename(columns = {'bp':'blood_pressure','sg':'specific_gravity',
'al':'albumin','su':'sugar','rbc':'red_blood_cells','pc':'pus_cell','pcc':'pus_cell_clumps','ba':'bacteria','bgr':'blood_glucose_random',
'bu':'blood_urea','sc':'serum_creatinine','sod':'sodium','pot':'potassium','hemo':'hemoglobin','pcv':'packed_cell_volume','wc':'white_blood_cell_count','rc':'red_blood_cell_count',
'htn':'hypertension','dm':'diabetes_mellitus','cad':'coronary_artery_disease','appet':'appetite',
'pe':'pedal_edema','ane':'anemia'}, inplace = True) 
```

```python
#check the dataset to make sure the headers are descriptive
original_data_df.head()
```

```python
# make sure dtype for all columns are numeric so that the algorithm works properly
original_data_df.info()
```

```python
# what are the unique values in the numeric and categorical columns
#Listed unique values in the original_data_df
numerical_features = []
categorical_features = []

for i in original_data_df.columns:
    if original_data_df[i].nunique()>7:
        numerical_features.append(i)
    else:
        categorical_features.append(i)
```

```python
print('Numerical features: ', numerical_features)
print('\nCategorical features: ', categorical_features)
```

```python
for feats in categorical_features:
    print(f'{feats} has {original_data_df[feats].unique()} categories.\n')
```

```python
#Replace incorrect values
original_data_df['diabetes_mellitus'] = original_data_df['diabetes_mellitus'].replace(to_replace = {'\tno':'no','\tyes':'yes',' yes':'yes'})
original_data_df['coronary_artery_disease'] = original_data_df['coronary_artery_disease'].replace(to_replace = '\tno', value='no')
original_data_df['classification'] = original_data_df['classification'].replace(to_replace = 'ckd\t', value = 'ckd')
```

```python
# check to make sure the incorrect values have been replaced
for feats in categorical_features:
    print(f'{feats} has {original_data_df[feats].unique()} categories.\n')
```

```python
for feats in numerical_features:
    print(f'{feats} has {original_data_df[feats].unique()} numeric_values.\n')
```

```python
#Replace incorrect values
original_data_df['white_blood_cell_count'] = original_data_df['white_blood_cell_count'].replace(to_replace = {'\t?':'','\t6200':'6200','\t8400':'8400'})
original_data_df['red_blood_cell_count'] = original_data_df['red_blood_cell_count'].replace(to_replace = '\t?', value='')
original_data_df['packed_cell_volume'] = original_data_df['packed_cell_volume'].replace(to_replace = '\t?', value = '')
```

```python
# check to make sure the incorrect values have been replaced
for feats in numerical_features:
    print(f'{feats} has {original_data_df[feats].unique()} numeric_values.\n')
```
